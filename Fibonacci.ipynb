```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
series = [0,1,1]
n = int(input("Enter number of terms : "))

num1 = 0
num2 = 1
num3 = num2
i = 1

while i <= n :
    i += 1
    num1, num2 = num2, num3
    num3 = num1 + num2
    series.append(num3)
print(series)
```

```python
plt.xlabel('Number of term')
plt.ylabel('Fib number')
plt.plot(series, marker='*', markersize='5', markerfacecolor='red')
plt.grid()
plt.tight_layout()
plt.title('Fibonacci Numbers')
```
